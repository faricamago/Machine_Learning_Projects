# Data Preprocessing

### a) with LabelEncoder()

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le = le.fit(sy_train)
y_train = le.transform(sy_train)
sy_train,y_train
```

### b) Processing attributes with MinMaxScaler()

```python
dfX_train.to_numpy()
```

```python
n1 = preprocessing.MinMaxScaler()
n1 =n1.fit(dfX_train.to_numpy())
X_train = n1.transform(dfX_train.to_numpy())
X_train
```

# Building Machine Learning Model for Classification

```python
from sklearn.neighbors import KNeighborsClassifier
kValue = []
knnScoreTest = []
knnScoreTrain = []
for y in range(1,76):
    knn = KNeighborsClassifier(n_neighbors = y)
    knn = knn.fit(X_train, y_train)
    y_test = le.transform(sy_test.to_numpy())
    X_test = n1.transform(dfX_test.to_numpy())
    kValue.append(y)
    knnScoreTest.append(knn.score(X_test, y_test))
    knnScoreTrain.append(knn.score(X_train, y_train))
knnScoreTrain
```
